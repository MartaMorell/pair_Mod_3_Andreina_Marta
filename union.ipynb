```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
data_sellers = pd.read_csv("insight.csv", index_col=0) # leer un archivo
data_sellers.shape
```

```python
data_rating = pd.read_csv("rating-and-performance.csv", index_col=0) #leer otro archivo
data_rating.shape
```

remombrar las dos de rating

Unid los dos DataFrames en uno solo, usando el método que considereis más correcto. Querremos quedarnos con todos los productos que estén en los dos conjuntos de datos.

```python
data_rating_renamed = data_rating.add_prefix('extra_')
```

```python
# Paso 2: concatenar horizontalmente si no hay columnns en comun o PK
df_final = pd.concat([data_sellers, data_rating_renamed], axis=1)

print(df_final.shape) #validar numero de columnas, cambio?
```

```python
df_final.columns
```

Explorad el nombre de las columnas, ¿hay alguna que esté repetida o que no nos aporte información de valor? ¿Hay alguna columna que solo tenga un valor? En caso de ser así, eliminadlas. NOTA Hay tres columnas que cuando exploramos sus valores únicos y sus frecuencias ayer, nos deberíamos haber dado cuenta que solo tiene un valor.

currnecy buyer, theme, crawl_month son iguales

```python
df_final.drop(["extra_currency_buyer", "extra_theme", "extra_crawl_month"], axis = 1, inplace=True)
df_final.head()
```

```python
print(df_final.columns)
```

Algunas de las columnas tienen . en sus nombres y en otras tenemos _. Unifica el nombre de las columnas para que todas tengan "_". Además, todas deben estar en minúsculas.

```python
nuevas_columnas = {columna: columna.lower().replace(".", "_") for columna in df_final.columns}
df_final.rename(columns=nuevas_columnas, inplace=True)
```

```python
df_final.columns
```

En la columna de merchant_info_subtitle, si nos fijamos hay algunas celdas donde tenemos entre paréntesis el número total de calificaciones para el comerciante (que es la misma información que la que tenemos en la columna "merchant_rating_count"). Eliminad de la columna "merchant_info_subtitle" la parte del string de los paréntesis. Guardar el resultado de vuestra limpieza en una columna que se llame igual, es decir, "merchant_info_subtitle".

```python
# Eliminar la parte entre paréntesis (y los paréntesis) de cada string
df_final['extra_merchant_info_subtitle'] = df_final["extra_merchant_info_subtitle"].str.replace(r'\s*\([^)]*\)', '', regex=True)


print(df_final)
```


Si nos fijamos, en la primera fila antes solo teníamos un paréntesis, pero los hemos eliminado en el paso anterior, y ahora tenemos un string vacío, lo cual no tiene mucho sentido. En este ejercicio debereis reemplazar esos strings vacios por nulos de Pandas (NOTA estos deben ser indicados de la siguiente forma np.nan, y deberéis importar numpy).

```python
df_final['new_extra_merchant_info_subtitle'] = df_final["extra_merchant_info_subtitle"].replace(r'^\s*$', np.nan, regex=True)
```

```python
df_final['new_extra_merchant_info_subtitle']
```
